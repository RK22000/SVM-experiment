#This is me attempting to visualize the idea of moving samples to a higer dimemsion for easier classification

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(42)
reds = np.random.normal(loc=-20, scale=5, size=50)
reds=np.append(reds, np.random.normal(loc=20, scale=5, size=40))

blues = (np.random.normal(loc=0, scale=5, size=30))
reds.shape
```

```python
plt.figure(figsize=(20,10))
plt.subplot(121)
plt.scatter(x=reds, y=[0]*len(reds),c=['red']*len(reds))
plt.scatter(x=blues, y=[0]*len(blues), c=['blue']*len(blues))
plt.subplot(122)
plt.scatter(x=reds, y=[i**2 for i in reds],c=['red']*len(reds))
plt.scatter(x=blues, y=[i**2 for i in blues], c=['blue']*len(blues))
minb = min(blues)
maxb = max(blues)
plt.plot([minb, maxb], [minb**2, maxb**2], c='black')
```

```python
plt.scatter(x=[0]*len(reds), y=[i**2 for i in reds],c=['red']*len(reds))
plt.scatter(x=[0]*len(blues), y=[i**2 for i in blues], c=['blue']*len(blues))
plt.title("Reduced feture")
```

# Observation
In the original 1d space we are not able to find a hyperdot that can seperate the reds and blues
so we elevate all samples to a 2d space where it is easy to find a hyper line that seperates the reds and blues

#Shift all reds and blues away from 0

```python
shift = 100
reds = reds + shift
blues = blues + shift
```

```python
plt.figure(figsize=(20,10))
plt.subplot(121)
plt.scatter(x=reds, y=[0]*len(reds),c=['red']*len(reds))
plt.scatter(x=blues, y=[0]*len(blues), c=['blue']*len(blues))
plt.subplot(122)
plt.scatter(x=reds, y=[i**2 for i in reds],c=['red']*len(reds))
plt.scatter(x=blues, y=[i**2 for i in blues], c=['blue']*len(blues))
minb = min(blues)
maxb = max(blues)
plt.plot([minb, maxb], [minb**2, maxb**2], c='black')
```

#Observation
The same data becomes harder to seperate even when elevated to a 2d space, if samples in the original space are shifted away from x=0
at this point is it a matter of finding being able to find a good hyper-line, or do we elevate this data to the 3rd dimension?

```python
fig = plt.figure(figsize=(15,13))
ax = fig.add_subplot(projection='3d')

ax.scatter(reds, [i**2 for i in reds], [i**3 for i in reds], marker="o", c='red')
ax.scatter(blues, [i**2 for i in blues], [i**3 for i in blues], marker="^", c='blue')
ax.plot3D(xs=[minb, maxb], ys=[minb**2, maxb**2], zs=[minb**3, maxb**3], c='black')
plt.show()
```
